```python
from numpy import random
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datapoint = 1000 
for dimension in [5, 2000, 5000, 10000]:
# dimention = 1000000 takes way too long
  x = random.rand(datapoint, dimension)


  distance = []

  for i in range(0,datapoint):
      for j in range(i+1,datapoint):
          distance.append(np.linalg.norm(x[i]-x[j]))




  ax = sns.histplot(data=distance, kde=True)
  ax.set_title('DataPoint : {} , Dimension : {}'.format(datapoint,dimension))
  ax.set_xlabel('Distance')
  plt.show()
```
