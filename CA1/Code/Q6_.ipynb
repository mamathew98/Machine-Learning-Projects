### Import Libraires

```python
import pandas as pd
import math
import random
```

### Import dataset

```python
# load the file and store it in dataset 
dataset = pd.read_csv('./Iris.csv')
dataset
```

### Transform class names to id

```python
def class2id(dataset):
    classes = dataset.Class.values.tolist()
    ids = []
    for i in range(len(classes)):
        if classes[i] not in ids:
            ids.append(classes[i])
    class_ids = []      
    for i in range(len(ids)):
        for j in range(len(classes)):
            if classes[j] == ids[i]:
                class_ids.append(i)
    dataset["Class_id"] = class_ids
    return dataset
```

```python
dataset = class2id(dataset)
dataset
```

### Transform Dataframe to id for easier usage
labels are saved in last element of array for easier use

```python
X = dataset[["Sepal_Length","Sepal_Width","Petal_Length","Petal_Width"]].values.tolist()
y = dataset.Class_id.values.tolist()
y
```

### Split data to train and test datasets

```python
def train_test_split(X, y, percentage):
    count = int(len(X) * percentage)
    X_train = []
    X_test = []
    y_train = []
    y_test = []
    X_test = list(X)
    y_test = list(y)
    while len(X_train) < count:
        index = random.randrange(len(X_test))
        X_train.append(X_test.pop(index))
        y_train.append(y_test.pop(index))
    return X_train, X_test, y_train, y_test
```

```python
ratio = 0.6
X_train, X_test, y_train, y_test = train_test_split(X, y, ratio)
print("TRAIN: \n", pd.DataFrame(X_train, columns=["Sepal_Length","Sepal_Width","Petal_Length","Petal_Width"]))
print("TEST: \n", pd.DataFrame(X_test, columns=["Sepal_Length","Sepal_Width","Petal_Length","Petal_Width"]))
print('Total number of examples are: ', len(X))
print('Out of these, training examples are: ', len(X_train))
print("Test examples are: ", len(X_test))
```

### Probabilistic functions

```python
def mean(numbers):
    return sum(numbers) / float(len(numbers))

def std_dev(numbers):
    avg = mean(numbers)
    variance = sum([pow(x - avg, 2) for x in numbers]) / float(len(numbers) - 1)
    return math.sqrt(variance)
 
def mean_StdDev(dataset):
    info = [(mean(attribute), std_dev(attribute)) for attribute in zip(*dataset)]
    return info
```

### Model Proibabilistic fitting

```python
def grouping(X, y):
      groups_dict = {}
      for i in range(len(X)):
          if (y[i] not in groups_dict):
              groups_dict[y[i]] = []
          groups_dict[y[i]].append(X[i])
      return groups_dict

def fit(X, y):
    info = {}
    dict = grouping(X, y)
    for classValue, instances in dict.items():
        info[classValue] = mean_StdDev(instances)
    return info
```

```python
model = fit(X_train, y_train)
model
```

### Gaussian Probability

```python
def gaussianProbability(x, mean, stdev):
    exp = math.exp(-(math.pow(x - mean, 2) / (2 * math.pow(stdev, 2))))
    return (1 / (math.sqrt(2 * math.pi) * stdev)) * exp
```

### Predict best choice class for given data

```python
def predict_class(model, X_test):
    probabilities = {}
    for classValue, classSummaries in model.items():
        probabilities[classValue] = 1
        for i in range(len(classSummaries)):
            mean, std_dev = classSummaries[i]
            x = X_test[i]
            probabilities[classValue] *= gaussianProbability(x, mean, std_dev)

    print(probabilities)
    
    bestLabel, bestProb = None, -1
    for classValue, probability in probabilities.items():
        if bestLabel is None or probability > bestProb:
            bestProb = probability
            bestLabel = classValue
    return bestLabel


def getPredictions(model, X_test):
    predictions = []
    for i in range(len(X_test)):
        result = predict_class(model, X_test[i])
        predictions.append(result)
    return predictions
```

```python
y_pred = getPredictions(model, X_test)
print("\nModel predictions are: \n", [int(x) for x in y_pred])
```

### Calculate accuracy of the model

```python
def accuracy_rate(y_test, y_pred):
    correct = 0
    for i in range(len(y_test)):
        if y_test[i] == y_pred[i]:
            correct += 1
    return (correct / float(len(y_test))) * 100.0

def confusion_mtx(y_test, y_pred):
    mtx = [ [ 0 for i in range(3) ] for j in range(3) ]
    for i in range(len(y_test)):
        mtx[int(y_test[i])][int(y_pred[i])] += 1
    return mtx
```

```python
accuracy = accuracy_rate(y_test, y_pred)
confusion = confusion_mtx(y_test, y_pred)
print("Accuracy of model is: ", accuracy)
print("Confusion of model is: \n", pd.DataFrame(confusion))
```

### OPTIMAL BAYES (SELF-DEVELOPED)

```python
import operator
```

```python
def mu(X, y):
    mu_dict = {}
    groups = grouping(X, y)
    for id, group in groups.items():
        mu_dict[id] = [np.mean(np.array(group), axis=0)]
    return mu_dict

def sigma(X, y):
    sigma_dict = {}
    groups = grouping(X, y)
    for id, group in groups.items():
        df = pd.DataFrame(group)
        cov = df.cov().to_numpy()
        sigma_dict[id] = cov
    return sigma_dict
```

```python
def optimal_fit(X_train, y_train):
    info = {}
    info["sigma"] = sigma(X_train, y_train)
    info["mu"] = mu(X_train, y_train)
    return info
```

```python
model = optimal_fit(X_train, y_train)
model
```

```python
def optimal_predict(model, vector):
    mus = model['mu']
    sigmas = model['sigma']
    best_label = -1
    best_probability = -1
    for id, mu_group in mus.items():
        det = np.linalg.det(sigmas[id]) 
        denominator = math.sqrt(2 * math.pi * det)
        sub = np.subtract(vector, mu_group)
        numerator = math.exp((-0.5)*np.matmul(np.matmul(sub, np.linalg.inv(sigmas[id])), sub.transpose()))
        f = numerator / denominator
        if f > best_probability:
            best_label = id
            best_probability = f
    return best_label
```

```python
def get_optimal_prediction(model, X_test):
    y_pred = []
    for vec in X_test:
        y_pred.append(optimal_predict(model, vec))
    return y_pred
```
