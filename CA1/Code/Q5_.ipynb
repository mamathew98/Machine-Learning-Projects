### IMAGE CLASSIFIER

### Import Libraries

```python
from matplotlib.pyplot import imread
import matplotlib.pyplot as plt
import pandas as pd
from PIL import Image
import numpy as np
import os
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

### Load images and transform them into Dataframe of variables
each RGB value in dataframe demonstrates mean of all pixels in image

```python
img_folder = r"F:\University\Master's\Machine Learning\Homeworks\HW1\Q5_Dataset\Images"
#listing files in images folder
list_img = os.listdir(img_folder)
img_array = []

#iterating over img_folder to get the images as arrays
for image in sorted(list_img):
    [file_name, ext] = os.path.splitext(image) #splitting file name from its extension
    arr = np.array(Image.open(os.path.join(img_folder, image))) #creating arrays for all the images
    [h, w] = np.shape(arr)[0:2]#calculating height and width for each image
    arr_dim = arr.ndim #calculating the dimension for each array
    arr_shape = arr.shape #calculating the shape for each array
    if arr_dim == 2:
        arr_mean = np.mean(arr)
        print(f'[{file_name}, greyscale={arr_mean:.1f}]')
    else:
        arr_mean = np.mean(arr, axis=(0,1))
        img_array.append([file_name, arr_mean[0], arr_mean[1], arr_mean[2]])
df = pd.DataFrame(img_array, columns=['Name', 'Red', 'Green', 'Blue'])
df
```

### Add Labels for each image and it's data

```python
for index, row in df.iterrows():
    if 'c' in row['Name']:
        df.loc[index,'Class'] = "Chelsea"
        df.loc[index,'Class_id'] = 1
    else:
        df.loc[index,'Class'] = "Manchester"
        df.loc[index,'Class_id'] = 0

df
```

### Split labels from data

```python
X = df[["Red","Green","Blue"]].values.tolist()
y = df.Class_id.values.tolist()
print("X: ", X)
print("y: ", y)
```

### Classify using multinomial naiive bayes (Supervised)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
clf = MultinomialNB()
y_pred = clf.fit(X_train, y_train).predict(X_test)
acc_score = accuracy_score(y_test, y_pred)
conf_mat = confusion_matrix(
        y_test, y_pred, labels = [0, 1])
pr_score = precision_score(y_test, y_pred)
re_score = recall_score(y_test, y_pred)
print("Accuracy of model is: ", acc_score)
print("Percision of model is: ", pr_score)
print("Recall of model is: ", re_score)
print("Confusion matrix of model is: \n", conf_mat)
```

### Classify using comparison of RED and BLUE values (Unsupervised)

```python
y_pred2 = []
for img in X:
    if img[0] > img[2]:
        y_pred2.append(0)
    else:
        y_pred2.append(1)

acc_score = accuracy_score(y, y_pred2)
conf_mat = confusion_matrix(
        y, y_pred2, labels = [0, 1])
pr_score = precision_score(y, y_pred2)
re_score = recall_score(y, y_pred2)
print("Accuracy of model is: ", acc_score)
print("Percision of model is: ", pr_score)
print("Recall of model is: ", re_score)
print("Confusion matrix of model is: \n", conf_mat)
```
